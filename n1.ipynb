```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. The Data (TV Budget vs Sales)
# We reshape X because the model expects a 2D array (a list of lists)
X = np.array([230.1, 44.5, 17.2, 151.5, 180.8]).reshape(-1, 1)
y = np.array([22.1, 10.4, 9.3, 18.5, 12.9])

# 2. The Model (The "Professional" Tool)
model = LinearRegression()
model.fit(X, y)  # This is where the model "learns" the trend

# 3. The Results
slope = model.coef_[0]       # How steep the line is
intercept = model.intercept_ # Where the line hits the y-axis

print(f"Model Equation: Sales = {slope:.3f} * TV_Budget + {intercept:.3f}")

# 4. Visualization
plt.scatter(X, y, color='blue', label='Actual Data (Noise)')
plt.plot(X, model.predict(X), color='red', label='Prediction (Signal)')
plt.xlabel('TV Budget ($k)')
plt.ylabel('Sales (Units)')
plt.legend()
plt.show()
```
