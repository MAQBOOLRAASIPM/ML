```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# 1. Generate dummy data
# Experience (X) and Salary (y)
X = np.array([[1.1], [1.3], [1.5], [2.0], [2.2], [2.9], [3.0], [3.2], [4.0], [5.0]])
y = np.array([39343, 46205, 37731, 43525, 39891, 56642, 60150, 54445, 64445, 67000])

# 2. Split Data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# 3. Train the Model
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# 4. View the learned parameters
print(f"Intercept (Beta 0): {regressor.intercept_:.2f}")
print(f"Slope (Beta 1): {regressor.coef_[0]:.2f}")

# 5. Make Predictions
y_pred = regressor.predict(X_test)

print("\nActual vs Predicted:")
for actual, pred in zip(y_test, y_pred):
    print(f"Actual: {actual}, Predicted: {pred:.2f}, Error: {actual - pred:.2f}")
```

```python
regressor.score(X_test,y_test)
```
